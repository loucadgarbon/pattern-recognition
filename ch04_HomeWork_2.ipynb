```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def generateY(x, w): 
    ϵ= np.random.normal( 
        loc=   0.0,  
        scale= np.random.uniform(1,10),  
        size= x.shape[1])     
    x= np.concatenate( 
            (np.ones((1,N)), 
             x),  
            axis=0) 
    y= w.T@x +ϵ 
    return y 
 

np.random.seed(0) 
N=    100 
 

low= -3.0 
high= 4.0 
b= -3  # y軸 截距 
 

w= np.array([b, 3, -2]) #.reshape(-1,1) 
 

x= np.random.uniform(low= low,  
                     high=high,  
                     size=(2,N)) 
 

y= generateY(x, w) 
 

%matplotlib inline
ax= pl.axes(projection='3d') 
ax.scatter3D(x[0,:],x[1,:],y, color='red') 
ax.set_xlabel('$x_0$') 
ax.set_ylabel('$x_1$') 
ax.set_zlabel('$y$') 
```

```python
from sklearn import linear_model 
 

# 要用 scikit-learn 的話， x 要轉置成 (100,2) 
xT= x.T 
 

# 然後，就一步到位！ 
reg= linear_model.LinearRegression() 
reg.fit(xT, y) 
 

# answer: 
reg.intercept_, reg.coef_ 
```

```python
x.shape
```

```python
pred_y = x[0,:]*reg.coef_[0] + x[1,:]*reg.coef_[1] + reg.intercept_
```

```python
%matplotlib inline
ax= pl.axes(projection='3d') 
ax.scatter3D(x[0,:],x[1,:],y, color='red') 
ax.scatter3D(x[0,:],x[1,:],pred_y, color='green') 
ax.set_xlabel('$x_0$') 
ax.set_ylabel('$x_1$') 
ax.set_zlabel('$y$') 
```
