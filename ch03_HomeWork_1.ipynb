```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

### Binomial Distribution

```python
percent_point = [0, 1/4, 2/4, 3/4, 1]
k = np.arange(0, 11)
X = st.binom(n=10, p=0.2).rvs(size=10)
pmf = st.binom(n=10, p=0.2).pmf(k)
cdf = st.binom(n=10, p=0.2).cdf(k)
ppf = st.binom(n=10, p=0.2).ppf(percent_point)
X, pmf , cdf, ppf
```

```python
fig = plt.plot(k, pmf)
```

```python
fig = plt.plot(k, cdf)
```

```python
st.binom(n=10, p=0.2).stats(moments='mvsk')
```
